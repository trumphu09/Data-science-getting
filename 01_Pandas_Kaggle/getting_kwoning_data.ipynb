```python
import pandas as pd

# %%
# 1. Tạo SERIES (Giống như tạo 1 cột danh sách món ăn)
# index là số thứ tự (0, 1, 2) mặc định
cot_mon_an = pd.Series(["Cơm", "Phở", "Bún"]) 
print("--- ĐÂY LÀ SERIES ---")
print(cot_mon_an)
print(type(cot_mon_an)) # Kết quả: pandas.core.series.Series

# %%
# 2. Tạo DATAFRAME (Giống như tạo cả menu)
# Dùng Dictionary để tạo: Khóa là tên cột, Giá trị là List
data = {
    "Tên Món": ["Cơm", "Phở", "Bún"],
    "Giá Tiền": [30 , 40, 35]
}
menu = pd.DataFrame(data)

print("\n--- ĐÂY LÀ DATAFRAME ---")
print(menu)
print(type(menu)) # Kết quả: pandas.core.frame.DataFrame

# %%
# 3. THỬ NGHIỆM: Lôi 1 cột từ DataFrame ra
cot_lay_ra = menu["Tên Món"]
print("\n--- Cột lấy ra từ DataFrame chính là Series ---")
print(type(cot_lay_ra)) # Mày sẽ thấy nó trả về Series
```

### Data Manipulation

```python
df.item_price.dtype
```

```python
# ham apply()
df. item_price = df.item_price.apply(lambda x : float(x.replace("$", "")))
```

```python
df.head()
```

```python
df.total_price = df.quantity * df.item_price
```

```python
df.head(6)
```

#### how much was the revenue for the period in the data set ?

```python
revenue = df.total_price.sum()
print(revenue)
```

#### Which was the most-ordered item ?

```python
# group by 
a = df.groupby("item_name")["quantity"].sum()
a.sort_values(ascending = False ).head(5)
```
