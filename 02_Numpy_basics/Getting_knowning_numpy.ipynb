## Broadcasting and Vectorized operations
Broadcasting is simply a set of a rules for appling  binary ufuncs (e.g. additon, subtraction, multiplication, etc.) on the arrays of different sizes.

```python
a = np.arange(0,10,2)
a
```

```python
a = a + 1 #Broadcasting 
a 
```

```python
a = np.array([1,2])
b = np.array([
    [2,3],
    [3,1]
])
c = a + b 
c
```

### Sorting along rows or columns

### Dot product example 

```python
# Number of jars sold
np.random.seed(0)
sales_amounts = np.random.randint(20,size = (5,3))
sales_amounts
```

```python
# Create weekly_sales DataFrame 
import pandas as pd
weekly_sales = pd.DataFrame(sales_amounts, index = ["Monday", "Tuesday","Wednesday", "Thursday", "Friday"], 
                            columns = ["Almond Butter", 'Peanut Butter', "Cashew Butter"])
weekly_sales
```

```python
# Create a price array 
prices = np.array([10,8,12])
butter_prices = pd.DataFrame(prices.reshape(1,3), index = ["Prices"], columns = ["Almond Butter", 'Peanut Butter', "Cashew Butter"])
```

```python
butter_prices
```

```python
res = weekly_sales @ butter_prices.T
```

```python
weekly_sales["Total_price"] = res 
weekly_sales
```

```python
np.sum(weekly_sales["Total_price"])
```
