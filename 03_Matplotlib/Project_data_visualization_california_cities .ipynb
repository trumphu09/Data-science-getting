```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
#import data 

cities = pd.read_csv("california_cities.csv")
cities.head(10)
```

```python
cities['area_total_km2'].sort_values(ascending = False)
```

```python
cities.shape
```

```python
# Extract latitude and longtitude
latitude = np.array(cities.latd)
longtitude = np.array(cities.longd)
```

```python
# coordinates of cities in california
fig, ax = plt.subplots()
ax.scatter(longtitude, latitude);
ax.set(title = 'cities in california ', xlabel = "longtitude", ylabel = "latitude" );
```

```python
#show how crowded and large this place is
fig, ax = plt.subplots()

popular , area =0.5*cities.population_total, 0.5*cities.area_total_km2
detail = ax.scatter(longtitude, latitude, s = area, c = popular );
ax.set(title = 'California Cities: Population and Area Distribution ', xlabel = "longtitude", ylabel = "latitude" );
fig.colorbar(detail);

ar = [50,100,500,800,1200]
for i in ar :
    plt.scatter([],[],s = i*0.5, c = 'k', alpha = 0.4, label = str(i) + "km$^2$")
plt.legend(labelspacing = 1);
```
