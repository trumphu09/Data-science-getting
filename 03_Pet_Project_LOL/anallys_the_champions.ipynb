```python
import pandas as pd 
import matplotlib as plt
```

```python
cham_lol = pd.read_csv("LoL_champions.csv")
cham_lol.head(6)
```

```python
cham_lol.describe()
```

```python
#how many unique champion in LoL 
print(cham_lol.Name.nunique())

cham_lol.Name.unique()
```

```python
cham_lol.shape
```

```python
# missing values
cham_lol.isnull().sum()
```

```python
# missing data in resourse tpye column
print(cham_lol["Resourse type"])
cham_lol = cham_lol.fillna(0)
```

```python
# information of yasuo
cham_lol.loc[cham_lol.Name == "Yasuo"]
```

```python
# How many champions have more than 600 base_hp ?
cham_lol.loc[cham_lol["Base HP"] > 600].Name.count()
```

```python
# How many champions in each different lane ?
cham_lol.Role.value_counts()
```

```python
# different role 
cham_lol.Tags.value_counts()
```

```python
cham_lol.columns = cham_lol.columns.str.replace(' ', '_')
```

```python
# Which has a higher average base armor: melee or ranged champions?
cham_lol.groupby("Range_type").Base_armor.mean()
```

```python
#top 5 cham has highest base AD
top5 = cham_lol.sort_values(by = "Attack_damage",ascending = False ).head(5)

'''
By selecting one of the top 5 champions with the highest base attack damage in LoL, 
players can gain a significant early-game advantage and dominate the laning phase.'''
```

```python
top5.plot(kind='bar', color='red', title='Top Champions by Damage')
```

```python
# in assassin's world 
As = cham_lol.loc[cham_lol.Tags == "Assassin"]
As
```

```python
As.sort_values(by = "Movement_speed",ascending = False)
```

```python
# highest HP at late 
cham_lol['Hp_level_18'] = cham_lol['Base_HP'] + cham_lol['HP_per_lvl'] * 17
cham_lol.Hp_level_18.sort_values(ascending = False )
cham_lol.sort_values(by = "Hp_level_18", ascending = False)
```

```python
# ass melee or mage range has higher mov speed
ass_melle = cham_lol.groupby(["Tags", "Range_type"]).Movement_speed.mean().sort_values(ascending = False)
ass_melle
```

```python
#Advanced screening
cham_lol.loc[ (cham_lol['Tags'].isin(["Fighter"])) & (cham_lol.Range_type == "Melee")  & (cham_lol.Attack_damage > 60) ]
```
