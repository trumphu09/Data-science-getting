### Pyplot API and Object-Orented API 
*  pyplot API - quickly
* OOP API

#### pyplot API 

```python
# Pyplot API
x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x),color = "green", linestyle = "dashed",label = 'sinx');
plt.plot(x,np.cos(x), label = "cosx");

plt.title("Comparing Sin and Cos")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend();
```

#### xlim, ylim --- axis 

```python
x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x));
plt.xlabel("x")
plt.ylabel("sin(x)")

plt.xlim(0,6)
plt.ylim(-1,0.5)
# xlim(a,b), ylim(c,d) <=> axis(a,b,c,d)
```

```python
x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x));
plt.xlabel("x")
plt.ylabel("sin(x)")

plt.axis([0,6,-1,0.5]);
```

#### Objiect-obriented API

```python
#1. prepare data 
x = [1,2,3,4]
y = [4,4,54,2]

#2. setup plat 
fig , ax = plt.subplots(figsize=(10,5)) # figure size = Width & Height of the plot 

#3. Plot the data 
plt.plot(x,y);

#4. Customize the plot
ax.set(title = "A simple plot", 
        xlabel = "x_axis",
        ylabel = "y_axis");
```
