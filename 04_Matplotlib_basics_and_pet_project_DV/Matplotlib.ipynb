### Pyplot API and Object-Orented API 
*  pyplot API - quickly
* OOP API

#### pyplot API 

```python
# Pyplot API
x = np.linspace(0,10,1000) # create 1000 numbers from 0 to 10
plt.plot(x,np.sin(x),color = "green", linestyle = "dashed",label = 'sinx'); #draw sin func 
plt.plot(x,np.cos(x), label = "cosx"); # cos func

plt.title("Comparing Sin and Cos")
plt.xlabel("x") # labeling to x columns
plt.ylabel("f(x)") # labeling to y columns
plt.legend();
```

#### xlim, ylim --- axis 

```python
x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x));
plt.xlabel("x")
plt.ylabel("sin(x)")

plt.xlim(0,6)
plt.ylim(-1,0.5)
# xlim(a,b), ylim(c,d) <=> axis(a,b,c,d)
```

```python
x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x));
plt.xlabel("x")
plt.ylabel("sin(x)")

plt.axis([0,6,-1,0.5]);
```

#### Objiect-obriented API

```python
#1. prepare data 
x = [1,2,3,4]
y = [4,4,54,2]

#2. setup plat 
fig , ax = plt.subplots(figsize=(10,5)) # figure size = Width & Height of the plot 

#3. Plot the data 
ax.plot(x,y);

#4. Customize the plot
ax.set(title = "A simple plot", 
        xlabel = "x_axis",
        ylabel = "y_axis");
```

```python
x = np.random.seed(2)
x = np.random.randint(0,100,[30])
y = np.random.uniform(10,500,[30]) # random 30 float num from 10 to 500

fig, ax = plt.subplots()
ax.scatter(x,y);
```

```python
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes =100*rng.rand(100)

fig, ax = plt.subplots()
img1 = ax.scatter(x,y, s = sizes, c = colors , cmap = 'viridis', alpha = 0.8  ); # {cmap : type color, alpha : blur }

fig.colorbar(img1); # add a color bar
```

#### Bar chart 
* Vertical chart
* Horible chart

```python
# prepare Data from a python dictionary
beverage = {"coke" : 10 , "pepsi": 12, "sprite" : 9}
fig, ax = plt.subplots()
ax.bar(beverage.keys(), beverage.values());
ax.set(title = "prices of HOT beverage", xlabel = "Drink_name", ylabel = "prices $");
```

```python
fig, ax = plt.subplots()
key = list(beverage.keys())
value = list(beverage.values())
ax.barh(key, value); # cant use hashable data type
```

#### Histogram

```python
# prepare data
np.random.seed(32)
student_height = np.random.normal(170, 10, 250) #averange = 170, std =10, include 250 student
student_height[:10]
```

```python
# OO API
np.random.seed(42)
student_height = np.random.normal(170, 10, 250)

fig, ax = plt.subplots()
ax.hist(student_height, bins = 30); #bins to seperate into smaller 
```
