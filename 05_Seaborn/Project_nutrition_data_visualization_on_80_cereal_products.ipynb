### Nutrition data visualization on 80 cereal products

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data set 
cereal_df = pd.read_csv("cereal.csv")
cereal_df.head()
```

```python
new = cereal_df.loc[:,["name","rating"]].sort_values(by = "rating")
```

```python
sns.barplot(data = new, x = "name", y = "rating")
```

```python
sns.boxplot(data = cereal_df.sort_values(by = "calories"), x = 'calories', y = 'sugars');
```

```python
cereal_new = cereal_df.loc[:,["calories","protein", "fat","sodium",'fiber','carbo','sugars','potass','vitamins']]
cereal_new
```

```python
sns.pairplot(data = cereal_new, hue = "calories")
```

```python
cereal_corr = cereal_new.corr() # get correlation data 
cereal_corr
```

```python
fig, ax = plt.subplots(figsize = (10,8))
sns.heatmap(data = cereal_corr, cmap = "Blues",  annot = True);
ax.set(title = "Correlation matrix sample cereals cooomposition");
```
